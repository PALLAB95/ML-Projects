# First Innings Score Prediction

```python
#Importing essential Libraries
import pandas as pd
import pickle

#Loading the Dataset
df = pd.read_csv("ipl.csv")
df.head()
```

```python
# -- Data Cleaning --
# Dropping the unwanted column
Unwanted_column =['mid','venue','batsman','bowler','striker','non-striker']
df.drop(labels = Unwanted_column, axis = 1 , inplace=True)
df.head()
```

```python
df['bat_team'].unique()
```

```python
#Keeping only consistent team
consistent_teams = ['Kolkata Knight Riders','Chennai Super Kings','Rajasthan Royals','Mumbai Indians',
                   'Kings XI Punjab','Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
# Removing the First 5 overs data in every Match
df = df[df['overs']>=5.0]
df.head()
```

```python
print(df['bat_team'].unique())
print(df['bowl_team'].unique())
```

```python
# Converting the column 'data' from string into datetime object
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x , '%Y-%m-%d'))
```

```python
# --- Data Preprocessing ---
# Converting Categorical Features useing OnehotEncoding method
encoded_df = pd.get_dummies(data = df, columns = ['bat_team','bowl_team'])
encoded_df.head()
```

```python
# Rearranging the columns
encoded_df = encoded_df[['date','bat_team_Chennai Super Kings',
       'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
       'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians',
       'bat_team_Rajasthan Royals', 'bat_team_Royal Challengers Bangalore',
       'bat_team_Sunrisers Hyderabad', 'bowl_team_Chennai Super Kings',
       'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
       'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians',
       'bowl_team_Rajasthan Royals', 'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad', 'runs', 'wickets', 'overs', 'runs_last_5',
       'wickets_last_5', 'total']]
```

```python
# Splitting The Data into train and test set
X_train = encoded_df.drop(labels='total',axis = 1)[encoded_df['date'].dt.year <= 2016]
X_test = encoded_df.drop(labels='total',axis = 1)[encoded_df['date'].dt.year >= 2017]
y_train = encoded_df[encoded_df['date'].dt.year <=2016 ]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >=2017 ]['total'].values
```

```python
# Removing The Date Column
X_train.drop(labels ='date',axis = True, inplace =True)
X_test.drop(labels ='date',axis = True, inplace =True)
```

```python
# --- Model Building ---
# Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Creating a pickle file for the classifier
filename = 'first-innings-score-lr-model.pkl'
pickle.dump(regressor, open(filename, 'wb'))
```
